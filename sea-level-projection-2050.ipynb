游늷 3. Se utiliza la funci칩n linregress del m칩dulo scipy.stats para calcular la pendiente y la intersecci칩n en el eje y de la l칤nea de mejor ajuste. Posteriormente, se grafica esta l칤nea sobre el gr치fico de dispersi칩n, extendi칠ndola hasta el a침o 2050 para predecir el incremento del nivel del mar en dicho a침o.

```python
# Importar bibliotecas necesarias
import scipy  # Biblioteca principal para an치lisis estad칤stico y cient칤fico
from scipy import stats  # Subm칩dulo de Scipy que incluye herramientas estad칤sticas avanzadas
from scipy.stats import linregress  # Funci칩n espec칤fica para realizar regresiones lineales
```

```python
df.columns
```

```python
# Crear el gr치fico de dispersi칩n
plt.figure(figsize=(15, 5))  # Definir el tama침o del gr치fico
sns.scatterplot(x='Year', y='CSIRO Adjusted Sea Level', data=df, 
                hue='CSIRO Adjusted Sea Level', # Diferenciar puntos con una escala de colores
                palette='Blues',  # Usar una paleta azul para claridad
                alpha=0.8 # Ajustar transparencia de los puntos
                )

# Calcular la l칤nea de mejor ajuste utilizando regresi칩n lineal
slope, intercept, r_value, p_value, std_error = linregress(df['Year'], df['CSIRO Adjusted Sea Level'])


# Crear valores del eje x para proyectar la l칤nea de mejor ajuste hasta 2050
years_extended = np.arange(1880,2051)
sea_levels_predicted = intercept + slope * years_extended

# Graficar la l칤nea de mejor ajuste
sns.lineplot(x=years_extended, y=sea_levels_predicted, color = 'red', 
             label='Best Fit Line', # Etiqueta para la l칤nea
             linewidth = 2, # Grosor de la l칤nea para mayor visibilidad
             )

# Agregar t칤tulos y etiquetas a los ejes
plt.title('Evoluci칩n del Nivel del Mar Ajustado por CSIRO con L칤nea de Mejor Ajuste', fontsize=15)
plt.xlabel('A침os', fontsize=12)
plt.ylabel('Nivel del Mar Ajustado por CSIRO (pulgadas)', fontsize=12)

# Mostrar la leyenda y mejorar el dise침o del gr치fico
plt.legend(title='CSIRO Adjusted Sea Level', fontsize=10)
plt.grid(True, alpha=0.5)  # Agregar una cuadr칤cula ligera para facilitar la lectura
plt.tight_layout()  # Optimizar el dise침o
plt.show()

# Predicci칩n del nivel del mar para el a침o 2050
sea_level_2050 = intercept + slope * 2050
print(f'El nivel del mar proyectado para el a침o 2050 es: {sea_level_2050:.2f} pulgadas.')
```

游닇An치lisis del Gr치fico:

El gr치fico muestra un incremento sostenido en el nivel del mar ajustado por CSIRO desde 1880 hasta una proyecci칩n para el a침o 2050. Los datos hist칩ricos evidencian un aumento moderado hasta mediados del siglo XX, seguido de una aceleraci칩n m치s notable en d칠cadas recientes. La l칤nea de mejor ajuste, calculada mediante regresi칩n lineal, proyecta un nivel del mar de aproximadamente 10.18 pulgadas para 2050.

Esta tendencia resalta el impacto continuo del cambio clim치tico en el aumento del nivel del mar, subrayando la importancia de implementar medidas de mitigaci칩n y adaptaci칩n, especialmente en regiones costeras vulnerables. Si bien el modelo lineal generaliza adecuadamente la tendencia hist칩rica, no considera posibles aceleraciones futuras que podr칤an intensificar a칰n m치s este fen칩meno.

游늷 4. Se grafica una nueva l칤nea de mejor ajuste utilizando 칰nicamente los datos desde el a침o 2000 hasta el 칰ltimo a침o disponible en el conjunto de datos. Adem치s, se extiende esta l칤nea hasta el a침o 2050 para proyectar el incremento del nivel del mar si la tasa de aumento observada desde el a침o 2000 contin칰a en el futuro.

```python
# Filtrar los datos a partir del a침o 2000
df_2000 = df[df['Year'] >= 2000]

# Crear el gr치fico de dispersi칩n
plt.figure(figsize=(15,5))
sns.scatterplot(x='Year', y='CSIRO Adjusted Sea Level', data=df_2000, 
                hue ='CSIRO Adjusted Sea Level', # Escalar colores seg칰n los valores de nivel del mar
                palette = 'Blues',  # Utilizar  una paleta azul para mayor claridad
                alpha = 0.8  # Ajustar la transparencia de los puntos
                )


# Calcular la l칤nea de mejor ajuste para los datos desde el a침o 2000
slope, intercept, r_value, p_value, std_error = linregress(df_2000['Year'], df_2000['CSIRO Adjusted Sea Level'])

# Crear los valores del eje x para extender la l칤nea de mejor ajuste hasta 2050
years_extended = np.arange(2000,2051)
sea_levels_predicted = intercept + slope * years_extended  # Calcular los valores predichos

# Graficar la l칤nea de mejor ajuste
sns.lineplot(x=years_extended, y=sea_levels_predicted, color='red',
             label = 'Best Fit Line (2000 onwards)', # Etiqueta descriptiva para la l칤nea
             linewidth = 2, # Grosor de la l칤nea para mayor visibilidad
             )

# Agregar t칤tulos y etiquetas
plt.title("Evoluci칩n del Nivel del Mar Ajustado por CSIRO desde el A침o 2000", fontsize=15)
plt.xlabel('A침os', fontsize=12)
plt.ylabel('Nivel del Mar Ajustado por CSIRO (pulgadas)', fontsize=12)

# Mostrar la leyenda y mejorar el dise침o del gr치fico
plt.legend(title='CSIRO Adjusted Sea Level', fontsize=10)
plt.grid(True, alpha=0.5)  # Mostrar una cuadr칤cula ligera para facilitar la lectura
plt.tight_layout()  # Optimizar el dise침o para evitar superposiciones
plt.show()

# Predicci칩n del nivel del mar en el a침o 2050 utilizando solo datos desde 2000
sea_level_2050 = intercept + slope * 2050

print(f"El nivel del mar proyectado para el a침o 2050 seg칰n CSIRO ser치: {sea_level_2050:.2f} pulgadas")
```

La proyecci칩n utilizando 칰nicamente datos recientes indica que el nivel del mar alcanzar치 aproximadamente 15.38 pulgadas para el a침o 2050, un valor significativamente mayor que la predicci칩n basada en datos desde 1880 (10.18 pulgadas).
