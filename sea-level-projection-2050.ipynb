Este proyecto analiza y pronostica el aumento del nivel del mar a nivel global hasta el año 2050, utilizando datos históricos desde 1880 y modelos de regresión lineal. Se examinan tendencias a largo plazo y patrones recientes de aceleración en el incremento del nivel del mar, destacando la influencia del cambio climático.

📌 2. Se utiliza la biblioteca Matplotlib para crear un gráfico de dispersión(scatter), empleando la columna 'Year' como eje x y la columna 'CSIRO Adjusted Sea Level' como eje y.

```python
# Creando el gráfico de dispersión
plt.figure(figsize=(15,5))
sns.scatterplot(x='Year', y='CSIRO Adjusted Sea Level', data=df, hue='CSIRO Adjusted Sea Level' ,palette='Reds', alpha = 0.8)

#Agregar títulos y etiquetas
plt.title('Evolución del Nivel del Mar Ajustado por CSIRO', fontsize = 15)
plt.xlabel('Año', fontsize=12)
plt.ylabel('Nivel del Mar Ajustado por CSIRO (pulgadas)', fontsize=12) 

#Ajustar diseño del gráfico
plt.grid(True, alpha = 0.5)
plt.tight_layout()
plt.show()
```

📝Análisis del gráfico:

Los puntos más altos y oscuros hacia los años recientes destacan la intensificación del fenómeno, lo que subraya la urgencia de medidas de mitigación y adaptación para las zonas costeras vulnerables.

📌 3. Se utiliza la función linregress del módulo scipy.stats para calcular la pendiente y la intersección en el eje y de la línea de mejor ajuste. Posteriormente, se grafica esta línea sobre el gráfico de dispersión, extendiéndola hasta el año 2050 para predecir el incremento del nivel del mar en dicho año.

```python
# Importar bibliotecas necesarias
import scipy  # Biblioteca principal para análisis estadístico y científico
from scipy import stats  # Submódulo de Scipy que incluye herramientas estadísticas avanzadas
from scipy.stats import linregress  # Función específica para realizar regresiones lineales
```

```python
df.columns
```

```python
# Crear el gráfico de dispersión
plt.figure(figsize=(15, 5))  # Definir el tamaño del gráfico
sns.scatterplot(x='Year', y='CSIRO Adjusted Sea Level', data=df, 
                hue='CSIRO Adjusted Sea Level', # Diferenciar puntos con una escala de colores
                palette='Blues',  # Usar una paleta azul para claridad
                alpha=0.8 # Ajustar transparencia de los puntos
                )

# Calcular la línea de mejor ajuste utilizando regresión lineal
slope, intercept, r_value, p_value, std_error = linregress(df['Year'], df['CSIRO Adjusted Sea Level'])


# Crear valores del eje x para proyectar la línea de mejor ajuste hasta 2050
years_extended = np.arange(1880,2051)
sea_levels_predicted = intercept + slope * years_extended

# Graficar la línea de mejor ajuste
sns.lineplot(x=years_extended, y=sea_levels_predicted, color = 'red', 
             label='Best Fit Line', # Etiqueta para la línea
             linewidth = 2, # Grosor de la línea para mayor visibilidad
             )

# Agregar títulos y etiquetas a los ejes
plt.title('Evolución del Nivel del Mar Ajustado por CSIRO con Línea de Mejor Ajuste', fontsize=15)
plt.xlabel('Años', fontsize=12)
plt.ylabel('Nivel del Mar Ajustado por CSIRO (pulgadas)', fontsize=12)

# Mostrar la leyenda y mejorar el diseño del gráfico
plt.legend(title='CSIRO Adjusted Sea Level', fontsize=10)
plt.grid(True, alpha=0.5)  # Agregar una cuadrícula ligera para facilitar la lectura
plt.tight_layout()  # Optimizar el diseño
plt.show()

# Predicción del nivel del mar para el año 2050
sea_level_2050 = intercept + slope * 2050
print(f'El nivel del mar proyectado para el año 2050 es: {sea_level_2050:.2f} pulgadas.')
```

📝Análisis del Gráfico:

El gráfico muestra un incremento sostenido en el nivel del mar ajustado por CSIRO desde 1880 hasta una proyección para el año 2050. Los datos históricos evidencian un aumento moderado hasta mediados del siglo XX, seguido de una aceleración más notable en décadas recientes. La línea de mejor ajuste, calculada mediante regresión lineal, proyecta un nivel del mar de aproximadamente 10.18 pulgadas para 2050.

Esta tendencia resalta el impacto continuo del cambio climático en el aumento del nivel del mar, subrayando la importancia de implementar medidas de mitigación y adaptación, especialmente en regiones costeras vulnerables. Si bien el modelo lineal generaliza adecuadamente la tendencia histórica, no considera posibles aceleraciones futuras que podrían intensificar aún más este fenómeno.

📌 4. Se grafica una nueva línea de mejor ajuste utilizando únicamente los datos desde el año 2000 hasta el último año disponible en el conjunto de datos. Además, se extiende esta línea hasta el año 2050 para proyectar el incremento del nivel del mar si la tasa de aumento observada desde el año 2000 continúa en el futuro.

```python
# Filtrar los datos a partir del año 2000
df_2000 = df[df['Year'] >= 2000]

# Crear el gráfico de dispersión
plt.figure(figsize=(15,5))
sns.scatterplot(x='Year', y='CSIRO Adjusted Sea Level', data=df_2000, 
                hue ='CSIRO Adjusted Sea Level', # Escalar colores según los valores de nivel del mar
                palette = 'Blues',  # Utilizar  una paleta azul para mayor claridad
                alpha = 0.8  # Ajustar la transparencia de los puntos
                )


# Calcular la línea de mejor ajuste para los datos desde el año 2000
slope, intercept, r_value, p_value, std_error = linregress(df_2000['Year'], df_2000['CSIRO Adjusted Sea Level'])

# Crear los valores del eje x para extender la línea de mejor ajuste hasta 2050
years_extended = np.arange(2000,2051)
sea_levels_predicted = intercept + slope * years_extended  # Calcular los valores predichos

# Graficar la línea de mejor ajuste
sns.lineplot(x=years_extended, y=sea_levels_predicted, color='red',
             label = 'Best Fit Line (2000 onwards)', # Etiqueta descriptiva para la línea
             linewidth = 2, # Grosor de la línea para mayor visibilidad
             )

# Agregar títulos y etiquetas
plt.title("Evolución del Nivel del Mar Ajustado por CSIRO desde el Año 2000", fontsize=15)
plt.xlabel('Años', fontsize=12)
plt.ylabel('Nivel del Mar Ajustado por CSIRO (pulgadas)', fontsize=12)

# Mostrar la leyenda y mejorar el diseño del gráfico
plt.legend(title='CSIRO Adjusted Sea Level', fontsize=10)
plt.grid(True, alpha=0.5)  # Mostrar una cuadrícula ligera para facilitar la lectura
plt.tight_layout()  # Optimizar el diseño para evitar superposiciones
plt.show()

# Predicción del nivel del mar en el año 2050 utilizando solo datos desde 2000
sea_level_2050 = intercept + slope * 2050

print(f"El nivel del mar proyectado para el año 2050 según CSIRO será: {sea_level_2050:.2f} pulgadas")
```

📝Análisis del Gráfico:

El gráfico muestra el incremento del nivel del mar ajustado por CSIRO desde el año 2000, evidenciando una aceleración en el ritmo de aumento comparado con las tendencias históricas completas. La proyección utilizando únicamente datos recientes indica que el nivel del mar alcanzará aproximadamente 15.38 pulgadas para el año 2050, un valor significativamente mayor que la predicción basada en datos desde 1880 (10.18 pulgadas).

Esta diferencia refleja un cambio acelerado en las últimas décadas, probablemente impulsado por el calentamiento global, el derretimiento de glaciares y la expansión térmica de los océanos. Los resultados subrayan la urgencia de implementar medidas para mitigar el cambio climático y planificar estrategias de adaptación, especialmente en áreas costeras vulnerables.

📌5. Se analizan los datos actualizados del nivel del mar proporcionados por NOAA (Administración Nacional Oceánica y Atmosférica) y se calcula la línea de mejor ajuste junto con la proyección del nivel del mar para el año 2050.

```python
df_NOAA = df.copy()
df_NOAA.head()
```

📌 Se eliminan los valores nulos del conjunto de datos para garantizar la integridad y consistencia en el análisis.

```python
# Eliminar valores nulos en la columna 'NOAA Adjusted Sea Level'
df_NOAA.dropna(subset=['NOAA Adjusted Sea Level'],inplace=True)
```

```python
# Crear el gráfico de dispersión
plt.figure(figsize=(15, 5))  # Establecer el tamaño del gráfico
sns.scatterplot(x='Year', y='NOAA Adjusted Sea Level', data=df_NOAA, 
            hue='NOAA Adjusted Sea Level',  # Escalar colores según los valores del nivel del mar
            palette='Blues',  # Usar una paleta azul para facilitar la interpretación
            alpha=0.8  # Ajustar la transparencia de los puntos
                )

# Calcular la línea de mejor ajuste con regresión lineal
slope, intercept,r_value,p_value, std_error = linregress(df_NOAA['Year'], df_NOAA['NOAA Adjusted Sea Level'])

# Crear los valores del eje x para proyectar la línea de ajuste hasta 2050
years_extended = np.arange(1993, 2051)  # Rango de años desde 1993 hasta 2050
sea_level_predicted = intercept + slope * years_extended  # Calcular los valores predichos para la línea

# Graficar la línea de mejor ajuste
sns.lineplot(x=years_extended, y=sea_level_predicted, color='red', 
            label='Best Fit Line',  # Etiqueta descriptiva para la línea
            linewidth=2  # Grosor de la línea para mayor visibilidad
             )

# Agregar títulos y etiquetas
plt.title('Nivel del Mar Ajustado por NOAA con Proyección a 2050', fontsize=15)
plt.xlabel('Años', fontsize=12)
plt.ylabel('Nivel del Mar Ajustado por NOAA (pulgadas)', fontsize=12)

# Mostrar leyenda y mejorar el diseño del gráfico
plt.legend(title='NOAA Adjusted Sea Level', fontsize=10)
plt.grid(True, alpha=0.5)  # Añadir una cuadrícula ligera para facilitar la lectura
plt.tight_layout()  # Ajustar el diseño para evitar superposiciones
plt.show()

# Predicción del nivel del mar para 2050 según NOAA
sea_level_2050 = intercept + slope * 2050

print(f'El nivel del mar proyectado para el año 2050 según NOAA será de: {sea_level_2050:.2f} pulgadas.')
```

📝Análisis del Gráfico

El gráfico muestra un aumento constante del nivel del mar ajustado por NOAA desde 1993, proyectando que alcanzará aproximadamente 12.56 pulgadas para el año 2050. La tendencia lineal evidencia un incremento gradual y predecible, lo que valida el uso de regresión lineal para este análisis.

📝Análisis general:

Las proyecciones de CSIRO y NOAA muestran un aumento continuo del nivel del mar hacia 2050, aunque con diferencias significativas. Mientras CSIRO proyecta un nivel de 15.38 pulgadas (con datos desde 2000), NOAA estima un valor menor de 12.56 pulgadas (con datos desde 1993). Estas discrepancias pueden deberse a diferencias en metodologías o periodos analizados, pero ambas confirman una tendencia alarmante de incremento constante.

En conjunto, los resultados subrayan la aceleración reciente del aumento del nivel del mar y la necesidad de medidas urgentes para mitigar sus impactos, especialmente en zonas costeras vulnerables. Utilizar múltiples fuentes refuerza la credibilidad y robustez de las conclusiones.
